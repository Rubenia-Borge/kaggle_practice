# THE PERCEPTRON ALGORITHM

# To understand Neural Networks, we must start studying the Perceptron. The Perceptron is the simplest form of artificial neural networks. Serves as a foundation to solve two-class classification problems.

```python
# The building blocks of the perceptron are the input values, weights, bias, weighted sum and activation function.
# The perceptron algorithm takes the input values, multiplies it by the weights and adds the bias. This results
# in a weighted sum that is passed through an activation function that returns an output value.

# Setup
import numpy as np
from numpy import array, random, dot
from random import choice
from pylab import ylim, plot
from matplotlib import pyplot as plt

# Input layer
data = [
    (array([0,0,0,0,0,1,1]), 0),
    (array([1,0,1,1,0,1,1]), 1),
    (array([1,1,0,0,1,0,1]), 1),
    (array([0,1,1,1,0,1,1]), 1),]

# Weights are initialized to random values
weights_vector = random.rand(7)
# Error is the difference between the value in the data and the value returned
# by the activation function
errors_vector = []
# lrate is a the learning rate, how fast we want to change the weights in each iteration
lrate = 0.001
# n is number of iterations that we will run the algorithm
n = 10

def update_weights(weights_vector, lrate, error, x):
    weights_vector = weights_vector + x*error*lrate
    
# This is the activation function. In this case, step_function
def step_function(x):
    if x < 0:
        return -1
    else:
        return 1

# Compute the error and update the weights
for i in range(n):
    x, y = choice(data)
    weighted_sum = dot(weights_vector, x)
    error = y - step_function(weighted_sum)
    errors_vector.append(error)    
    update_weights(weights_vector,lrate,error,x)

# Compute the weighted sum and output value   
for x, _ in data:
    weighted_sum = dot(x, weights_vector)
    print("Input: {} - Weighted Sum: {} - Output: {}".format(x[:6], weighted_sum, step_function(weighted_sum)))


plt.figure(figsize=(9, 3))
plot(errors_vector, color="red")
plt.title("Errors Vector Plot")
plt.xlabel("Iterations")
plt.ylabel("Error Value")
plt.show()
```
