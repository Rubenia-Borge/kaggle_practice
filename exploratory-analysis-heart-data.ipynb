```python
# Exploratory Data Analysis
# Heart Disease Dataset
# ML Tasks: Classification
# https://archive.ics.uci.edu/ml/datasets/Heart+Disease
```

```python
'''
The purpose of data analysis is to analyze and summarize the most important characteristics of 
datasets. Also, to gain a better understanding of the variables and the relationships
between them.

'''
```

```python
# Load the data
# Pandas is a software library for data manipulation in Python

# This line of code calls the Pandas library to be used in your code
import pandas as pd 
from matplotlib import pyplot as plt

# heart_data_1 is the name of the dataframe
# a dataframe is a container for data, it has two dimensions
# rows for data and colums for labels

heart_data_1 = pd.read_csv("/kaggle/input/heart1/heart.csv")
```

```python
# Now we are going to perform an exploratory data analysis
# One of the goals of a exploratory data analysis is to gain understanding of the 
# data, find patterns and relationships between the data

# Pandas has a set of functions that allow to perform exploratory data analysis, EDA
```

```python
# head()
# Returns the top five rows in the dataset by default
heart_data_1.head()
```

```python
# If you add a number inside the parenthesis
# it will return that number of rows starting from the top
heart_data_1.head(10)
```

```python
# tail()
# Returns the bottom five rows in the dataset by default
heart_data_1.tail()
```

```python
# If you add a number inside the parenthesis
# it will return that number of rows starting from the bottom
heart_data_1.tail(10)
```

```python
# info()
# It gives a summary of the dataframe
# It shows the number of non-null-values and
# data types

heart_data_1.info()
```

```python
# shape
# Shows the number of dimensions and size of each dimension
# dataframes have two dimensions, rows and columns
# series have only one dimension, like an array or list
heart_data_1.shape
```

```python
# size
# Returns the number of elements in the dataframe or series.
# if the object is a dataframe, it computes the number of elements as
# number of rows multiplied by number of columns
# if the object is a series, then the number of rows is the number of elements
# series have only rows
# dataframes have rows and columns
heart_data_1.size
```

```python
# ndim
# Returns the number of dimensions of an object
# Used to find out if the object is a dataframe or a series
# series have a dimension of 1
# dataframes have a dimension of 2
heart_data_1.ndim
```

```python
# describe()
# Returns a statistical summary of the numberical columns in the 
# dataframe
heart_data_1.describe()
```

```python
# sample()
# Allows to select a random sample of rows from the dataframe or series
heart_data_1.sample()
```

```python
# If we leave the parenthesis empty, it returns one row
# if we specify a number inside the parenthesis, it will return
# that number of rows
heart_data_1.sample(5)
```

```python
# isnull()
# Shows the elements that are null
heart_data_1.isnull()
```

```python
# isnull().sum()
# use the function sum() to find out the number of null elements in
# each column
heart_data_1.isnull().sum()
```

```python
# nunique()
# Returns the number of unique elements in each column
# It counts the number of unique elements in a row or colum

heart_data_1.nunique()
```

```python
# If you put the name of the column inside squared brackets
# it will give you the number of unique elements in 
# that specific column
heart_data_1["thal"].nunique()
```

```python
# Another example
heart_data_1["oldpeak"].nunique()
```

```python
# index
# Returns the first and last index numbers
heart_data_1.index
```

```python
# memory_usage()
# Returns how much memory each column uses in bytes
heart_data_1.memory_usage()
```

```python
# dropna()
# Removes rows and colums from a dataframe that contain
# NaN or missing values
heart_data_1.dropna()
```

```python
# nlargest()
# Returns the first n rows ordered by columns in descending order

heart_data_1.nlargest(1,"age")
```

```python
heart_data_1.nlargest(1,"heart_rate")
```

```python
heart_data_1.nlargest(3,"cholestoral")
```

```python
# isin()
# shows whether or not a range of values is contained in a dataframe
# a dataframe column or a series
heart_data_1["age"].isin([23,34,67,89]).head()
```

```python
# isna()
# Returns a dataframe with boolean values
# True indicates there is a missing value
heart_data_1.isna().head()
```

```python
# Returns a boolean series indicating duplicate rows.
# Finds duplicated rows and returns a series indicating those
# duplicated rows
heart_data_1.duplicated()
```

```python
# value_counts()
# Returns a series with the counts of unique values
# Can be used for each column of the dataframe

heart_data_1["st_segment"].value_counts()
```

```python
# Shows the pairwise correlation of all columns in the dataframe
# Missing values and non-numeric data types are ignored
# What's important is pairwise correlation between features and target variable
# and correlation between feature variables
heart_data_1.corr()
```

```python
# dtypes
# Shows the data type of each column
heart_data_1.dtypes
```
